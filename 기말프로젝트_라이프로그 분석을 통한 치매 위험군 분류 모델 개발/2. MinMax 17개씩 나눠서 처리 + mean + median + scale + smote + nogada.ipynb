```python
from sklearn.metrics import accuracy_score


# 메소드 만들어서 보관
def evaluate(model, xtrain, ytrain, xtest, ytest, xtrains, ytrains, xtrain_ss, xtest_ss, xtrain_rbs, xtest_rbs, xtrain_sss, xtest_sss, xtrain_rbss, xtest_rbss):
    import matplotlib.pyplot as plt
    
    #Size
    plt.figure(figsize=(12, 6))

    #SMOTE X No Scaling
    scores = []
    plt.subplot(2,3,1)
    for i in range(10,100,5):
        model = RandomForestClassifier(n_estimators=i, random_state=123)
        model.fit(xtrain, ytrain)
        att = model.predict(xtest)
        acc = accuracy_score(ytest, att)
        scores.append(acc)
        
        #print('>%d, acc: %.3f' %(i, acc))
    plt.plot(range(10, 100, 5), scores, 'b', label='SMOTE X No Scaling')
    plt.title('SMOTE X No Scaling')

    #SMOTE O No Scaling
    plt.subplot(2,3,4)
    scores = []
    for i in range(10,100,5):
        model = RandomForestClassifier(n_estimators=i, random_state=123)
        model.fit(xtrains, ytrains)
        att = model.predict(xtest)
        acc = accuracy_score(ytest, att)
        scores.append(acc)
    plt.plot(range(10, 100, 5), scores, 'b', label='SMOTE O No Scaling')
    plt.title('SMOTE O No Scaling')
    
    #SMOTE X Standard Scaling
    plt.subplot(2,3,2)
    scores = []
    for i in range(10,100,5):
        model = RandomForestClassifier(n_estimators=i, random_state=123)
        model.fit(xtrain_ss, ytrain)
        att = model.predict(xtest_ss)
        acc = accuracy_score(ytest, att)
        scores.append(acc)
    plt.plot(range(10, 100, 5), scores, 'b', label='SMOTE X Standard Scaling')
    plt.title('SMOTE X Standard Scaling')

    #SMOTE X Robust Scaling
    plt.subplot(2,3,3)
    scores = []
    for i in range(10,100,5):
        model = RandomForestClassifier(n_estimators=i, random_state=123)
        model.fit(xtrain_rbs, ytrain)
        att = model.predict(xtest_rbs)
        acc = accuracy_score(ytest, att)
        scores.append(acc)
    plt.plot(range(10, 100, 5), scores, 'b', label='SMOTE X Robust Scaling')
    plt.title('SMOTE X Robust Scaling')

    #SMOTE O Standard Scaling
    plt.subplot(2,3,5)
    scores = []
    for i in range(10,100,5):
        model = RandomForestClassifier(n_estimators=i, random_state=123)
        model.fit(xtrain_sss, ytrains)
        att = model.predict(xtest_sss)
        acc = accuracy_score(ytest, att)
        scores.append(acc)
    plt.plot(range(10, 100, 5), scores, 'b', label='SMOTE O Standard Scaling')
    plt.title('SMOTE O Standard Scaling')

    #SMOTE O Robust Scaling
    plt.subplot(2,3,6)
    scores = []
    for i in range(10,100,5):
        model = RandomForestClassifier(n_estimators=i, random_state=123)
        model.fit(xtrain_rbss, ytrains)
        att = model.predict(xtest_rbss)
        acc = accuracy_score(ytest, att)
        scores.append(acc)
    plt.plot(range(10, 100, 5), scores, 'b', label='SMOTE O Robust Scaling')
    plt.title('SMOTE O Robust Scaling')

    plt.tight_layout()
    plt.show()
    
    print()
```
